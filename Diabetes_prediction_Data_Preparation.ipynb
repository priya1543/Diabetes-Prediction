```python
data.drop(columns=['Income','Education','NoDocbcCost','AnyHealthcare'],inplace=True)
```

Feature Importance using Random Forest

```python
y = data.iloc[:,0:1] # dependent variable
x = data.iloc[:,1:] # dependent variables
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x, y)
importances = rf.feature_importances_
feature_importances = pd.DataFrame({'Feature': x.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)
selected_features_rf= feature_importances.head(16)['Feature']
X_selected_rf = x[selected_features_rf]
```

```python
plt.bar(x=feature_importances['Feature'], height=feature_importances['Importance'], color='#087E8B')
plt.title('Feature importances obtained from Random Forest', size=10)
plt.xticks(rotation='vertical')
plt.show()
```

Feature Importance using XGBoost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x, y)
importances = xgb.feature_importances_
feature_importances_xgb = pd.DataFrame({'Feature': x.columns, 'Importance': importances})
feature_importances_xgb = feature_importances_xgb.sort_values(by='Importance', ascending=False)
selected_features_xgb= feature_importances_xgb.head(16)['Feature']
X_selected_xgb = x[selected_features_xgb]
```

```python
plt.bar(x=feature_importances_xgb['Feature'], height=feature_importances_xgb['Importance'], color='purple')
plt.title('Feature importances obtained from XGBoost Classifier', size=10)
plt.xticks(rotation='vertical')
plt.show()
```

Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from feature_engine.selection import RecursiveFeatureElimination
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
RFE_model = RecursiveFeatureElimination(
    estimator = model,
    scoring = 'r2',
    threshold = 0.001,
    cv=2,
)
RFE_model.fit(x,y)
```

```python
RFE_model.feature_importances_.plot.bar(color="teal")
plt.ylabel('Feature importance')
plt.title('Feature importance from Logistic Regression',size=10)
plt.show()
```

The importance of features obtained for each model is different, and thus not removing any of the features.
